```python
# Train a logistic regression classifier to predict whether a flower is iris virginica or not
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
iris = datasets.load_iris()
```

```python
X = iris["data"][:, 3:]
y = (iris["target"] == 2)
```

```python
# Train a logistic regression classifier
clf = LogisticRegression()
clf.fit(X,y)
example = clf.predict(([[2.6]]))
print(example)
```

```python
# Using matplotlib to plot the visualization
X_new = np.linspace(0,3,1000).reshape(-1,1)  #linspace gives 1000 points between 0 and 3 #reshape changing the shape of data 
# print(X_new)
y_prob = clf.predict_proba(X_new)

print(y_prob)

```

```python

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.plot(X_new, y_prob[:,1], "g-", label="virginica")
plt.title("LOGISTIC REGRESSION", loc = 'center',fontdict = font1)
plt.ylabel("probability",fontdict = font2)
plt.xlabel("taken points",fontdict = font2)
plt.show()
```
