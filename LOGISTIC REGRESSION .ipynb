#    Train  a logistic  regression  clssifier to predict  whether a flower is iris virginic or not ! 

```python
from sklearn import datasets
import numpy as np 
from sklearn.linear_model import LogisticRegression 
import matplotlib as plt
import matplotlib.pyplot as  plt
```

```python
iris=datasets.load_iris()
print(iris)
```

```python
iris.data
```

```python
iris.keys()
```

```python
iris.target
```

```python
iris.data[:,3]#printing the perticular colunm form the dataset 
```

```python
x=iris.data[:,3:]#features (choosing one feature which is on 3rd index =petal width  )
x
y = (iris["target"] == 2)#labels (here using binary clssifier to classification) 2for virginia 
# y
#astype(np.int ) is the numpy's function which is used to give value in number 
```

TRAIN A LOGISTIC REGRESSION CLASSIFIERS 

```python
clf=LogisticRegression ()
clf.fit(x,y)
```

```python
example =clf.predict(([[1.6]]))

example
```

```python
example1 =clf.predict(([[2.6]]))
example1
```

USING MATPLOTLIB TO PLOT  THE VISUALIZATION

```python
x_new =np.linspace(0,3,1000).reshape(-1 ,1)
x_new
```

```python
y_prob=clf.predict_proba(x_new)
print(y_prob)
```

```python
plt.plot(x_new ,y_prob[:,1],"g-",label="virginica")
plt.show()
```
